Pipelining

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numeric_features = ['Adult Mortality','Alcohol']
categorical_features = ['Country','Status']
```

```python
# Numeric Transformer
numeric_transformer=Pipeline(
    steps=[
        ('imputation',SimpleImputer(strategy='median')),
        ('scaling',StandardScaler())
    ])
```

```python
# Categorical Transfomer
categorical_transformer=Pipeline(
    steps=[
        ('imputation',SimpleImputer(strategy='most_frequent')),
        ('encoding',OneHotEncoder(handle_unknown='ignore'))
    ])
```

```python
transformer=ColumnTransformer(
    transformers=[
        ('numeric_transformations',numeric_transformer, numeric_features),
        ('categorical_transformations',categorical_transformer,categorical_features)
    ])
```

```python
X_processed = transformer.fit_transform(life_df)
```
